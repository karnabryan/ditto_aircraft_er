```python
import torch
import os

# Check if CUDA is available to PyTorch
print("CUDA available:", torch.cuda.is_available())

# Print PyTorch version and the CUDA version PyTorch was built with
print("PyTorch version:", torch.__version__)
print("PyTorch CUDA version:", torch.version.cuda)

# Check the CUDA_HOME environment variable
print("CUDA_HOME:", os.environ.get("CUDA_HOME"))
```

```python
if torch.cuda.is_available():
    # Create a tensor and move it to GPU
    x = torch.rand(5, 5).cuda()
    print("Tensor on GPU:\n", x)
else:
    print("CUDA is not available.")
```

```python
if torch.cuda.is_available():
    # Create two random tensors and multiply them on GPU
    a = torch.rand(1000, 1000, device='cuda')
    b = torch.rand(1000, 1000, device='cuda')
    c = torch.mm(a, b)
    print("Matrix multiplication on GPU successful. Result shape:", c.shape)
else:
    print("CUDA is not available.")
```
